```python
from without_rl import without_reinforcement_learning

# Without reinforcement learning
num_of_tests = 1000
timesteps, penalties, rewards, rewards_per_step = [], [], [], []
for k in range(num_of_tests):
    clear_output(wait=True)
    print(f"Testing scenarij: {k+1}")
    f, t, p, r = without_reinforcement_learning()
    timesteps.append(t)
    penalties.append(p)
    rewards.append(r)
    rewards_per_step.append(r/t)

povprecna_nagrada = sum(rewards) / num_of_tests
povprecno_korakov = sum(timesteps) / num_of_tests
povprecno_kazni = sum(penalties) / num_of_tests
povprecna_nagrada_korak = sum(rewards_per_step) / num_of_tests

print ("Povprecna nagrada:", povprecna_nagrada)
print ("Povprecno stevilo korakov:", povprecno_korakov)
print ("Povprecno stevilo kazni na epizodo:", povprecno_kazni)
print ("Povprecna nagrada na korak:", povprecna_nagrada_korak)
```
